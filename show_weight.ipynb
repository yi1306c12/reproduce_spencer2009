```python
import numpy
import matplotlib.pyplot as plt
```

```python
data = numpy.fromfile('test_weights.dat', dtype=numpy.float32).reshape((1001, 1000))
weights = data[:1000, :]
weights_noise = data[1000, :]
```

```python
plt.matshow(numpy.fromfile('test_noise.dat', dtype=numpy.int16).reshape((1100, -1)), cmap='binary')
plt.xlabel('neuron ID')
plt.ylabel('time[us]')
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
heatmap = ax.matshow(numpy.fromfile('test_spk.dat', dtype=numpy.int16).reshape((1100, -1)).T, cmap='binary')
ax.xaxis.tick_bottom()
plt.xlabel('time[ms]')
plt.ylabel('neuron ID')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
heatmap = ax.matshow(numpy.fromfile('test_Vm.dat', dtype=numpy.float64).reshape((1100, -1)).T)
ax.xaxis.tick_bottom()
plt.xlabel('time[ms]')
plt.ylabel('neuron ID')
plt.colorbar(heatmap)
plt.show()
```
