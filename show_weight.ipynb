```python
fig, ax = plt.subplots()
heatmap = ax.matshow(numpy.fromfile(dirpath/(condname+'_spk.dat'), dtype=numpy.int16).reshape((1100, -1)), cmap='binary')
ax.xaxis.tick_bottom()
plt.xlabel('neuron ID')
plt.ylabel('time[ms]')
plt.show()
```

```python
fig, ax = plt.subplots()#figsize=(16, 10))
heatmap = ax.matshow(numpy.fromfile(dirpath/(condname+'_v.dat'), dtype=numpy.float64).reshape((1100, -1)))
ax.xaxis.tick_bottom()
plt.xlabel('neuron ID')
plt.ylabel('time[ms]')
plt.colorbar(heatmap)
plt.show()
```

```python
def conductance_plot(typename, n):
    fig, ax = plt.subplots()#figsize=(16, 10))
    heatmap = ax.matshow(numpy.fromfile(dirpath/(condname+'_'+typename+'.dat'), dtype=numpy.float32).reshape((-1, n)))
    ax.xaxis.tick_bottom()
    plt.title(typename)
    plt.xlabel('neuron ID')
    plt.ylabel('time[ms]')
    plt.colorbar(heatmap)
    plt.show()
import itertools
for name in itertools.product(['g_'], ['r_', 'd_'], ['ampa', 'nmda', 'gaba']):
    conductance_plot(''.join(name), 1000)
```
